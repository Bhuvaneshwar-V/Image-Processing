```python
import cv2
import glob
import matplotlib.pyplot as plt
```

```python
# storing the file_names in the folder using glob
file_list = glob.glob('Images/*.*')
print(file_list)
```

```python
# storing the file_names in the list using glob
my_list = []

path = 'Images/*.*'
for file in glob.glob(path):
    print(file)
    img = cv2.imread(file)
    my_list.append(img)
```

```python
# converting the BGR image to RGB image and displaying using matplotlib
img_RGB = cv2.cvtColor(my_list[0], cv2.COLOR_BGR2RGB)
plt.imshow(img_RGB)
```

```python
# reading and writing multiple images using glob
img_number = 1

for file in glob.glob(path):
    a = cv2.imread(file)
    cv2.imwrite('Images/color_images/'+'img'+str(img_number)+'.jpg', a)
    img_number += 1
```
